```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('full_data_flightdelay.csv')

# EDA: Distribution of delays
sns.histplot(data['DEP_DEL15'], bins=2, kde=False)
plt.title("Distribution of Flight Delays")
plt.show()

# Correlation heatmap
corr_matrix = data[['MONTH', 'DEP_DEL15', 'PRCP', 'SNOW', 'SNWD', 'TMAX', 'AWND', 'LATITUDE', 'LONGITUDE']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap (2 Decimal Points)")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score
from tqdm.notebook import tqdm
import pandas as pd

# Prepare data
X = data[['MONTH', 'PRCP', 'SNOW', 'SNWD', 'TMAX', 'AWND', 'LATITUDE', 'LONGITUDE']]
y = data['DEP_DEL15']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize model
model = LogisticRegression(max_iter=1000, random_state=42)

# Simulated progress bar for training
with tqdm(total=100, desc="Training Progress") as pbar:
    model.fit(X_train, y_train)
    pbar.update(100)

# Evaluate
y_pred = model.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
print(f"Precision: {precision_score(y_test, y_pred):.2f}")
print(f"Recall: {recall_score(y_test, y_pred):.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm

# Monte Carlo: Simulate 100 scenarios of weather conditions
simulations = 100
results = []

with tqdm(total=simulations, desc="Monte Carlo Simulations") as pbar:
    for _ in range(simulations):
        # Add random noise to the test data to simulate weather uncertainty
        X_simulated = X_test + np.random.normal(0, 0.1, size=X_test.shape)
        
        # Predict probabilities for simulated data
        y_simulated = model.predict_proba(X_simulated)[:, 1]  # Probability of delay
        
        # Store results
        results.append(y_simulated)
        
        # Update progress bar
        pbar.update(1)

# Aggregate results
mean_prob = np.mean(results, axis=0)
std_dev_prob = np.std(results, axis=0)

# Plot uncertainty
plt.errorbar(range(len(mean_prob)), mean_prob, yerr=std_dev_prob, fmt='o')
plt.title("Predicted Delay Probability with Uncertainty")
plt.xlabel("Test Sample Index")
plt.ylabel("Delay Probability")
plt.show()
```

```python
import numpy as np

# Aggregate results
mean_prob = np.mean(results, axis=0)
std_dev_prob = np.std(results, axis=0)

# Select a subset of test samples to plot
subset_size = 50  # Plot for only 50 samples to reduce clutter
indices = range(subset_size)
mean_subset = mean_prob[:subset_size]
std_dev_subset = std_dev_prob[:subset_size]

# Plot with shaded confidence intervals
plt.figure(figsize=(12, 6))
plt.plot(indices, mean_subset, label="Mean Probability", color="blue")
plt.fill_between(indices, 
                 mean_subset - std_dev_subset, 
                 mean_subset + std_dev_subset, 
                 color="blue", alpha=0.2, label="1 Std Dev")
plt.title("Predicted Delay Probability with Uncertainty (Shaded Confidence Interval)")
plt.xlabel("Test Sample Index (Subset)")
plt.ylabel("Delay Probability")
plt.legend()
plt.show()
```
